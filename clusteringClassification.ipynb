```python
import pandas as pd
from random import randint,uniform
```

```python
df = pd.read_csv("out.csv") # diubah kenumerik
dfcluster = df.drop('status', axis=1) #untuk clustering
```

```python
dfcluster
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(dfcluster[['suhu udara','kelembaban udara','suhu tanah','kelembaban tanah','cahaya']])
```

```python
print(kmeans.cluster_centers_)
```

```python
print(kmeans.labels_)
```

```python
clust = []
j = 0
for i in dfcluster[['suhu udara','kelembaban udara','suhu tanah','kelembaban tanah','cahaya']].values:
    clust.append([i[0],i[1],i[2],i[3],i[4],kmeans.labels_[j]])
    j+=1
dfclust = pd.DataFrame(clust,columns=['suhu udara','kelembaban udara','suhu tanah','kelembaban tanah','cahaya','status'])
```

```python
penyesuaian = [x for x in range(4)]
fitur = ['suhu udara','kelembaban udara','suhu tanah','kelembaban tanah','cahaya']
label = ['mendung siram', 'mendung kering','cerah siram','cerah kering']
for i in range(4):
    dekat = 1000000000
    for j in range(4):
        meanclust = dfclust[(dfclust['status']==i)].describe().loc['mean']
        meanclass = df[(df['status']==label[j])].describe().loc['mean']
        meanclust = meanclust[fitur[1]]+meanclust[fitur[4]]
        meanclass = meanclass[fitur[1]]+meanclass[fitur[4]]
        dekat1 = abs(meanclust-meanclass)
        print(i,j,dekat1)
        if dekat>dekat1:
            penyesuaian[i]=j
            dekat = dekat1
print(penyesuaian)
```

```python
label_hasil_clustering = []
for x in list(kmeans.labels_):
    label_hasil_clustering.append(penyesuaian[x])
```

```python
print(label_hasil_clustering)
print(list(df['status']))
```
