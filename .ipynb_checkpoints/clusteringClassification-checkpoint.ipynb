```python
import pandas as pd
from random import randint,uniform
```

```python
data = []
cols = ['a','b','c','d','y']
for i in range(500):
    y = randint(0,3)
    if y==0:
        a = randint(1,10)
        b = uniform(1,2)
        c = 10+(uniform(0,1)*10)
        d = 100+(uniform(0,1)*100)
    elif y==1:
        a = randint(11,20)
        b = uniform(3,4)
        c = 20+(uniform(0,1)*10)
        d = 200+(uniform(0,1)*100)
    elif y==2:
        a = randint(21,30)
        b = uniform(5,6)
        c = 30+(uniform(0,1)*10)
        d = 300+(uniform(0,1)*100)
    else:
        a = randint(31,40)
        b = uniform(7,8)
        c = 40+(uniform(0,1)*10)
        d = 400+(uniform(0,1)*100)
    data.append([a,b,c,d,y])
df = pd.DataFrame(data,columns=cols)
```

```python
df
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
kmeans = KMeans(n_clusters=4)
kmeans.fit(df[['a','b','c','d']])
```

```python
print(kmeans.cluster_centers_)
```

```python
print(list(kmeans.labels_))
```

```python
print(list(df['y']))
```

```python
penyesuaian = [x for x in range(4)]
for i in range(4):
    dekat = 100
    for j in range(4):
        dekat1 = abs(kmeans.cluster_centers_[i][0]-df[(df['y']==j)].describe().loc['mean']['a'])
        print(i,j,dekat1)
        if dekat>dekat1:
            penyesuaian[i]=j
            dekat = dekat1
print(penyesuaian)
```

```python
label_hasil_clustering = []
for x in list(kmeans.labels_):
    label_hasil_clustering.append(penyesuaian[x])
```

```python
print(label_hasil_clustering)
print(list(df['y']))
```

```python
salah = 0
label_asli = list(df['y'])
for i in range(len(label_hasil_clustering)):
    if label_hasil_clustering[i] != label_asli[i]:
        salah += 1
print(salah)
print(500-salah)
```
