```python
data = []
for i in range(1,100,1):
    data.append([i,i*randint(1,3),i*randint(1,5),i*randint(1,2)])
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
sel = SelectKBest(chi2,k=3).fit_transform(data,label)
```

```python
sel.shape
```

```python
print(sel)
```

```python
from sklearn.feature_selection import SelectPercentile
```

```python
sel = SelectPercentile(chi2,percentile=90).fit_transform(data,label)
```

```python
from sklearn.feature_selection import SelectFpr
```

```python
sel = SelectFpr(chi2, alpha=1).fit_transform(data,label)
```

```python
from sklearn.feature_selection import GenericUnivariateSelect
```

```python
transformer = GenericUnivariateSelect(chi2, mode='fpr', param=0.01)
```

```python
sel = transformer.fit_transform(data,label)
```

```python
for i in range(1,500,1):
    sel = GenericUnivariateSelect(chi2, mode='fpr', param=i/100).fit_transform(data,label)
    print(i/100,sel.shape)
```

```python
for i in range(1,500,1):
    sel = GenericUnivariateSelect(chi2, mode='fdr', param=i/100).fit_transform(data,label)
    print(i/100,sel.shape)
```

```python
for i in range(1,500,1):
    sel = GenericUnivariateSelect(chi2, mode='fwe', param=i/100).fit_transform(data,label)
    print(i/100,sel.shape)
```

```python
from sklearn.feature_selection import mutual_info_classif,mutual_info_regression
```

```python
label = []
for i in range(1,100,1):
    if i<30:
        label.append('sedikit')
    elif i<70:
        label.append('sedang')
    else:
        label.append('banyak')
```

<em>n_neighbors</em> dalam <em>mutual_info_classif</em> membandingkan apa?

```python
sel = mutual_info_classif(data,label,n_neighbors=3)
```

```python
print(sel)
```

Apakah pada <em>mutual_info_regression</em> input data bisa berupa string?

```python
sel = mutual_info_regression(data,label,n_neighbors=3)
```

```python
label = []
for i in range(1,100,1):
    if i<30:
        label.append(randint(1,10))
    elif i<70:
        label.append(randint(11,20))
    else:
        label.append(randint(21,30))
```

```python
sel = mutual_info_regression(data,label,n_neighbors=3)
```

```python
print(sel)
```
