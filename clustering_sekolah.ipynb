```python
import pandas as pd
from random import randint,uniform
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('datasekolah.csv')
```

```python
print(df)
```

```python
data = df[["Kabupaten","Asal_Sekolah","jurusan"]]
print(data)
```

<em>n_clusters</em> pada k-means clustering menandakan apa?

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(data)
```

```python
print(kmeans.labels_)
```

```python
silscore = silhouette_score(data,kmeans.labels_)
print(silscore)
```

```python
ax = sns.scatterplot(x=df["Kabupaten"],y=df["Asal_Sekolah"],hue=kmeans.labels_)
```

```python
ax = sns.scatterplot(x=df["jurusan"],y=df["Asal_Sekolah"],hue=kmeans.labels_)
```

```python
ax = sns.scatterplot(x=df["jurusan"],y=df["Kabupaten"],hue=kmeans.labels_)
```
