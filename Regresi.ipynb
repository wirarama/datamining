```python
import pandas as pd
from random import randint
```

```python
data = []
for x in range(1000):
    data.append([x,x*2,x*100,x*(-100)])
```

```python
data = []
for x in range(1000):
    data.append([x*randint(1,9),x*randint(100,999),(x*randint(100,999))/100,x*-1*randint(20,99)])
```

```python
data
```

```python
df = pd.DataFrame(data,columns=['a','b','c','d'])
```

```python
df
```

```python
data = df[['a','b','c']]
```

```python
label = df['d']
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(data, label, test_size=0.10, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg = DecisionTreeRegressor(max_depth=5)
reg.fit(train_x,train_y)
```

```python
y = reg.predict(test_x)
```

```python
y
```

```python
test_y
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
mean_squared_error(test_y,y)
```

```python
mean_absolute_error(test_y,y)
```

```python
regr_1 = DecisionTreeRegressor(max_depth=2)
regr_2 = DecisionTreeRegressor(max_depth=5)
regr_1.fit(train_x,train_y)
regr_2.fit(train_x,train_y)
y_1 = regr_1.predict(test_x)
y_2 = regr_2.predict(test_x)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
plt.scatter(train_x['b'].values,train_y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(test_x['b'].values, y_1, color="cornflowerblue", label="max_depth=2", linewidth=2)
plt.plot(test_x['b'].values, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```
